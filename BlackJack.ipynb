```python
#Deck
# Returns 2 cards randomly from the deck
from random import shuffle
class Deck:
    
    suits = ['Hearts', 'Diamonds' , 'Clubs' , 'Spades']
    ranks = ['Two' , 'Three' , 'Four' , 'Five' , 'Six' , 'Seven' , 'Eight' , 'Nine', 'Ten' , 'Ace', 'Jack', 'King' ,'Queen']
    
    def __init__(self,check):
        self.check = check
        
    def shuffle_deck_2(self ,cards):

        shuffle(Deck.ranks)
        shuffle(Deck.suits)
        
        i =0
        while True: 
            if ( [Deck.ranks[i] , Deck.suits[i]] in cards ) and ([Deck.ranks[i+1] , Deck.suits[i+1]] in cards ) :
                i+=1
                continue
            else:
                cards_on_deck = [[Deck.ranks[i] , Deck.suits[i]] , [Deck.ranks[i+1] , Deck.suits[i+1]] ]
                return cards_on_deck
    
    def shuffle_deck_1(self , cards):
        
        shuffle(Deck.ranks)
        shuffle(Deck.suits)
        
        i = 0
        while True:
            if [Deck.ranks[i] , Deck.suits[i]] in cards:
                i+=1
                continue
            else:
                cards_on_deck = [Deck.ranks[i] , Deck.suits[i]]
                return cards_on_deck


    #Prints out the deck on the table 
    def display_1(self ,cards):
        print(f"{cards[0][0]} of {cards[0][1]}")
                    
                    
    def display(self,cards):
        for i in range(0,len(cards)):
            print(f"{cards[i][0]} of {cards[i][1]}")
                
```

```python
# find the sum of the card points

def Sum_of_cards(cards):
    ace = False
    d = {"One":1 , "Two": 2 , "Three": 3 ,"Four": 4, "Five": 5, "Six": 6 , "Seven": 7, "Eight": 8, "Nine" : 9 , "Ten": 10 , "Jack":10 , "Queen": 10 ,"King": 10 , "Ace":11}

    sum = 0
    count =0
    for i in range(0,len(cards)):
        sum = sum + d[cards[i][0]]
        
        if cards[i][0] == "Ace":
            count+=1
            
            
    if sum >21 and count>0:
        result = Ace_check(sum ,count)
        return result

    else:
        return sum
```

```python
#if an Ace is present it can be counted as 1 or 11
def Ace_check(sum,count):
    while sum>21:
        if count==0:
            break
        else:
            sum = sum - 10
            count-=1
            continue
            
    return sum
```
